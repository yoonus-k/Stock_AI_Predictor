```python
# add the parent directory to the system path
from pathlib import Path
import sys
import os
current_dir = Path(os.getcwd())
main_dir = str(current_dir.parent.parent)
sys.path.append(main_dir)

from Sentiment.API.cot_API import get_cot_data

df = get_cot_data("XAUUSD")
df
```

```python
# save as csv
df.to_csv("XAUUSD_cot_data.csv", index=True)
```

```python
# plot the net non-commercial positions and the net non_reportable positions
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['Net_Noncommercial'], label='Net Non-Commercial Positions', color='blue')
plt.plot(df.index, df['Net_Nonreportable'], label='Net Non-Reportable Positions', color='orange')
plt.title('COT Data for NASDAQ')
```
